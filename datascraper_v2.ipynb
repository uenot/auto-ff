```python
STARTING_YEAR = 2019
ENDING_YEAR = 2015

season_dicts = {}

for year in range(STARTING_YEAR, ENDING_YEAR, -1):
    response = requests.get(f'http://www.fantasyfootballdatapros.com/api/players/{year}/all')
    player_data = response.json()
    current_players = []
    for player in player_data:
        # checks if player with matching name is in the current season
        if year != STARTING_YEAR:
            if player['player_name'] in [n['player_name'] for n in season_dicts[STARTING_YEAR]]:
                current_players.append(player)
        else:
            current_players.append(player)

    
    for player in current_players:
        # flattens "stats" dict into each individual stat
        for stat_category in player['stats'].values():
            for name, value in stat_category.items():
                player[name] = value
        del player['stats']
        

    season_dicts[year] = current_players
```
